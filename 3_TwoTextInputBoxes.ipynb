```python
import pygame

HEIGHT = 720
WIDTH = 1280

#Global constants here
BLACK = (255,255,255)
BLACK = (0,0,0)
GREY  = (50,50,50)
RED  = (207,0,0)

systemExit=False
pygame.init()
clock = pygame.time.Clock()
screen = pygame.display.set_mode((WIDTH, HEIGHT), 0 ,32)
base_font = pygame.font.Font(None,32)
color_active = pygame.Color('lightskyblue3')
color_passive = pygame.Color('gray15')

input_text1 = ''
input_text2 = ''
input_rect2 = pygame.Rect(200,300,140,32)
input_rect1 = pygame.Rect(200,200,140,32)
active_input_1 = False
active_input_2 = False

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            systemExit=True;
            break
        if event.type == pygame.MOUSEBUTTONDOWN:
            if input_rect1.collidepoint(event.pos):
                active_input_1 = True
                active_input_2 = False
            elif input_rect2.collidepoint(event.pos):
                active_input_1 = False
                active_input_2 = True
            else:
                active_input_1 = False
                active_input_2 = False
        if event.type == pygame.KEYDOWN:
            if active_input_1 == True:
                if event.key == pygame.K_BACKSPACE:
                    input_text1 = input_text1[:-1]
                elif event.key == pygame.K_RETURN:
                    active_input_1 = False
                    active_input_2 = True
                else:
                    input_text1 += event.unicode
            elif active_input_2 == True:
                if event.key == pygame.K_BACKSPACE:
                    input_text2 = input_text2[:-1]
                elif event.key == pygame.K_RETURN:
                    active_input_1 = True
                    active_input_2 = False
                else:
                    input_text2 += event.unicode
                    
    if systemExit==True:
        break
        
    screen.fill((0,0,0))
    
    if active_input_1:
        pygame.draw.rect(screen,color_active,input_rect1,2)
        pygame.draw.rect(screen,color_passive,input_rect2,2)
    elif active_input_2:
        pygame.draw.rect(screen,color_passive,input_rect1,2)
        pygame.draw.rect(screen,color_active,input_rect2,2)
    else:
        pygame.draw.rect(screen,color_passive,input_rect1,2)
        pygame.draw.rect(screen,color_passive,input_rect2,2)    
    
    # Render Text
    text_surface1 = base_font.render(input_text1,True,(255,255,255))
    text_surface2 = base_font.render(input_text2,True,(255,255,255))
    # Show Text inside Rectangle
    screen.blit(text_surface1, (input_rect1.x+5,input_rect1.y+5))
    screen.blit(text_surface2, (input_rect2.x+5,input_rect2.y+5))
    # Make Rectangle grow with Text
    input_rect1.w = max(100,text_surface1.get_width() + 10)
    input_rect2.w = max(100,text_surface2.get_width() + 10)
    
    pygame.display.flip()
    clock.tick(60)

exit()
```
